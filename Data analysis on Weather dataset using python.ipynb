find all the instances where weather is clear or visibility is above 40

```python
weather_data[(weather_data['Weather condition']=='Clear') | (weather_data['Visibility_km']>40)]
```

Find all instances where weather is clear and relative humidity is greater than 50 or visibility is above 40

```python
weather_data[((weather_data['Weather condition']=='Clear') & (weather_data['Rel Hum_%']>50)) | (weather_data['Visibility_km']>40)]
```
